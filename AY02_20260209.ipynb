## Operadores
Los operadores permiten trabajar sobre las variables, a la manera de las operaciones matemáticas.

* Operadores aritméticos

Los operadores aritméticos son diréctamente operaciones matemáticas estándar. Y se respeta la jerarquía de operaciones.
            
Operadores  | Aritméticos
------------|------------------
`a + b`     | suma
`a - b`     | Resta
`a * b`     | Multiplicación
`a / b`     | Diviisión
`a // b`    | División entero mínimo
`a % b`     | Módulo (resto de la división entera)
`a ** b`    | Potencia
` - a `     | Negativo


* Operadores de asignaciones

Los operadores simplifican operadores aritméticos comúnes.

Operadores | de asignaciones
-----------|------------------
`a += b`   | a = a + b
`a -= b`   | a = a - b
`a *= b`   | a = a * b
`a /= b`   | a = a / b
`a //= b`  | a = a // b
`a %= b`   | a = a % b
`a **= b`  | a = a ** b

* Operadores de comparación

Los comparadores dan resultados **lógicos** (si/no, true/false)

Operadores | de comparación
-----------|------------------
`a == b`   | a igual a b
`a != b`   | a distinto de b
`a < b`    | a menor a b
`a > b`    | a mayor a b
`a <= b`   | a  igual o menor que b
`a >= b`   | a mayor o igual que b


* Operadores de identidad y pertenencia

Operadores de | identidad y pertenencia
--------------|------------------
`a is b`      | a es el mísmo objeto que b
`a is not b`  | a no es el mísmo objeto que b
` a in b`     | a está contenido en b
`a not in b`  | a no está contenido en b

## Variables

En todos los lenguajes de programación, el nombre de las variables representan un valor dado alojado en la memoria. El valor puede cambiar, pero no el tipo de variable. Sin embargo, esto no pasa en Python ya que las variables son dinámicas.

Una variable puede tener un nombre corto (como x e y) o un nombre más descriptivo (edad, nombre_vehículo, volumen_total).

Reglas para el nombre de las variables de Python:

- Deben comenzar con una letra o el carácter de guión bajo
- No pueden comenzar con un número
- Solo pueden contener caracteres alfanuméricos y guiones bajos (Az, 0-9 y _)
- Los nombres de las variables distinguen entre mayúsculas y minúsculas (nombre, Nombre y NOMBRE son tres variables diferentes)
- Un nombre de variable no puede ser ninguna de las palabras clave de Python .

Con = asignamos el valor a una variable (Creamos en la memoria la variable), con == hacemos la comparación de dos cosas (idéntico a ). Nótese que = no es una operación simétrica.

En Python, las variables actúan como etiquetas o referencias a los objetos de datos en la memoria, en lugar de contenerlos directamente. Este sistema de 'punteros' es fundamental para la eficiencia del lenguaje, optimizando el uso de la memoria y la velocidad al manipular información.

Hay variables **globales** (se pueden usar o llamar en todo el notebook) y variables **locales** (se definen en una parte específica del código, como una función, y sólo se pueden llamar en dicha sección, se llaman localmente.)

Al crear una variable se asigan cierto espacio en la memoria a esa variable. Para los enteros son 32 bits, float son 64 bits y para string son 128 bits.


### Palabras reservadas

## Ejercicio 3

Diseña un programa que pida el valor de los tres lados de un triángulo y que calcule el valor de su área y perímetro. Puedes usar la fórmula de Herón de Alejandría.

La fórmula de Herón para calcular el área ($A$) de un triángulo, dados sus tres lados $a, b, c$, es la siguiente:

$$A = \sqrt{s(s-a)(s-b)(s-c)}$$

Donde $s$ es el semiperímetro del triángulo, calculado como:

$$s = \frac{a+b+c}{2}$$

````python
print('Programa que calcula el área y perimetro de un triangulo')
a=float(input('Dime el lado a '))
b=float(input('Dime el lado b '))
c=float(input('Dime el lado c '))

perimetro=a+b+c
print('El perimetro del triangulo es:', perimetro)
s= (perimetro)/2
area= (s*(s-a)*(s-b)*(s-c))**0.5
print('El area del triangulo es', area)
````
